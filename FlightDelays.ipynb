```python
# Loading the 3rd file of dataset: flights.csv
pd.set_option('display.max_columns',None)
df3 = pd.read_csv('flights.csv',header=0)
df3
```

**Domain Analysis:**<break>
* YEAR, MONTH, DAY and DAY_OF_WEEK: Represents date, day, month, week and year of the airline making a trip.<break>
* AIRLINE, FLIGHT_NUMBER, TAIL_NUMBER: Airline with respective flight number and tail number.<break>
* ORIGIN_AIRPORT, DESTINATION_AIRPORT: Airline's trip origin and destination airport.<break>
* SCHEDULED_DEPARTURE, DEPARTURE_TIME, DEPARTURE_DELAY, SCHEDUED_ARRIVAL, ARRIVAL_TIME, ARRIVAL_DELAY: Scheduled time of departure and arrival, Actual time of departure and arrival and if there is any deay in departure and arrival.<break>
* TAXI_OUT, WHEELS_OFF, WHEELS_ON, TAXI_IN: Time of taxing out, taxing in, wheels off and wheels in.<break>
* SCHEDULED_TIME, ELAPSED_TIME: Time scheduled for take off and actual time taken.<break>
* AIR_TIME: Amount of time the airplane is in air.<break>
* DISTANCE: Distance covered by airplane from source to destination.<break>
* DIVERTED, CANCELLED, CANCELLATION_REASON: If the airplane was diverted/cancelled and if so the reason for cancellation.<break>
* AIR_SYSTEM_DELAY, SECURITY_DELAY, AIRLINE_DELAY, LATE_AIRCRAFT_DELAY, WEATHER_DELAY: If there was any delay due to security, airline, aircraft or weather reasons.
